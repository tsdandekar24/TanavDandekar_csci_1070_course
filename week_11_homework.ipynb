```python
import pandas as pd
import numpy as np

marketing_df = pd.read_csv("marketing_campaign.csv", sep = "\t")
marketing_df.head(20)
```

```python
marketing_df["Income"] = marketing_df["Income"].fillna(marketing_df["Income"].median())
marketing_df.head(10)
```

I found nulls in the Income column, so I replaced nulls with the median income.

```python
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()
import datetime as dt

# Converting education and marital_status into numerical values
marketing_df["Education"] = class_labels.fit_transform(marketing_df["Education"].values)
marketing_df["Marital_Status"] = class_labels.fit_transform(marketing_df["Marital_Status"].values)

# Converted the date format into a float
#marketing_df["Dt_Customer"] = pd.to_datetime(marketing_df["Dt_Customer"], format = "%d-%m-%Y")
#marketing_df["Dt_Customer"] = marketing_df["Dt_Customer"].apply(lambda x: x.timestamp())

# Dropped Z_CostContact and Z_Revenue because all of the ID's had the same values for them (3 and 11 respectively)
#marketing_df = marketing_df.drop(columns = ["Z_CostContact", "Z_Revenue"], axis = 1)

# Found the sum of all the mnt products each customer bought and created a new column for the total
mnt_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
marketing_df['Sum_Mnt'] = marketing_df[mnt_columns].sum(axis = 1)

marketing_df.head(20)
```

# #2

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist


X = marketing_df.drop("Response", axis = 1)
y = marketing_df["Response"]

def k_means(r1, r2, dist_metric):
    inertias = []
    distortions = []

    for k in range(r1, r2):
        model = KMeans(n_clusters = k).fit(X)

        distance = cdist(X, model.cluster_centers_, dist_metric)
        distortion = sum(np.min(distance, axis = 1)) / X.shape[0]
        distortions.append(distortion)

        inertia = model.inertia_
        inertias.append(inertia)
k_means(1, 10, "euclidean")
```

```python
def plt_distortions(r1, r2):
    plt.plot(range(r1, r2), distortions)
    plt.title("Elbow Method with Distortion")
plt_distortions(1, 10)
```

```python
def plt_inertia(r1, r2):
    plt.plot(range(r1, r2), inertias)
    plt.title("Elbow Method with Inertia")
plt_inertia(1, 10)
```

I used the elbow method to graph the distortion and inertias. The bend in the graph help shows which number of clusters may work welel. According to the graph, it looks like the number of clusters that would best fit the dataset is 3 to 4.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(X, "ward")
plt.figure(figsize = (10, 7))

dendro = dendrogram(linked, labels = list(y))
plt.xlabel("Customers")
plt.ylabel("Date Time")
```

To find the number of clusters, I looked at the dendrogram and counted the groups of data.

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters = 4, linkage = "ward")
cluster.fit_predict(X)
```
