```python
import pandas as pd
import numpy as np

belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
group_belize_df = belize_df.groupby("neighbourhood")
price_df = group_belize_df[["price"]].mean().reset_index()
price_df
```

```python
sum_df = group_belize_df[["number_of_reviews"]].sum().reset_index()
sum_df
```

```python
max_df = group_belize_df[["availability_365"]].max().reset_index()
max_df
```

```python
small_df = pd.merge(price_df, sum_df, on = "neighbourhood")
final_df = pd.merge(small_df, max_df, on = "neighbourhood")
final_df
```
