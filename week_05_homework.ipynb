# Question #1


When using machine learning, it is best to use it when you are trying to make models from actual data. Based on the data given and the model, predictions can be made about the data. Machine learning can be used when trying to predict things like fraud detection, forecasting the weather, and targeted marketing. All of them use previous data from past models to predict what will happen or what is happening. For example, when forecasting the weather, the previous data about temperatures, wind changes, and how sunny it is can help models predict what will be weather be today, tomorrow, and next week. Even though the weather predictions are not always right, the prediction models are usually accurate. However, there are cases when machine learning is not needed. A situation where machine learning is not helpful is when the data set is not large enough. If the data set is not large enough, it will be hard to make predictions because there could be some bias. The data set could be slightly skewed compared to a sufficient data set where there is some correlation. Another example where machine learning is not needed is when rules can be used to sort the data. When I mean rule based, there can be certain conditions to fit the data. For example, an if-then statement could be used. If a person is a certain age, then they are part of a certain group. In this situation, the data types are discrete and categorical. Categories data types typically do not do well with machine learning. 

# Question #2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()

titanic_df = pd.read_csv("titanic.csv")
titanic_df.head(10)
```

Making numerical values to sex. Male = 1 and Female = 0

```python
titanic_df["Sex"] = class_labels.fit_transform(titanic_df["Sex"].values)
titanic_df.head()
```

Using one-hot encoding on the Embarked column

```python
embarked_df = pd.get_dummies(titanic_df[["Embarked"]])
titanic_df = pd.concat((titanic_df, embarked_df), axis = 1)
titanic_df.head()
```

Substituting null values of age with the median age of the dataset and putting NAN cabins with Unknown

```python
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df[["Age"]].median())
titanic_df["Cabin"] = titanic_df["Cabin"].fillna("Unknown")
titanic_df.head(10)
```

# Question #3

```python
titanic_cor = titanic_df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]]
titanic_cor.corr()
```

```python
import seaborn as sns
sns.heatmap(titanic_cor.corr(), annot = True)
```

Based on the correlation matrix, there is a slight negative correlation between sex and survived. The negative correlation indicates that more women survived than men did. There is a slight positive correlation between age and fare. Generally, the older the person is the more expensive the fare was. Also, there is a slight negative correlation between passenger class and age. The older the person means the higher passenger class they were in.

```python
import time
time1 = time.perf_counter()
age_bin = titanic_df["Age"].apply(lambda x: "Children" if x <= 12 else "Teenager" if x <= 19 else "Adult" if x <=64  else "Senior")
titanic_df["age_bin"] = age_bin
time2 = time.perf_counter()
print(time2 - time1)
```

```python
time1 = time.perf_counter()
ages = [0, 12, 19, 64, 100]
labels = ["Children", "Teenager", "Adult", "Senior"]
titanic_df["age_bin"] = pd.cut(titanic_df["Age"], ages, labels = labels)
time2 = time.perf_counter()
print(time2 - time1)
```

It takes less time to use a lamda function and set it values than it takes to bin all of the data. This may be because it takes time to look at the ages and labels list and assign it to the age_bin column. Also, I feel like there is less code for the lambda function than it is to bin it.

```python
child_df = titanic_df.loc[titanic_df["age_bin"] == "Children"]
teen_df = titanic_df.loc[titanic_df["age_bin"] == "Teenager"]
adult_df = titanic_df.loc[titanic_df["age_bin"] == "Adult"]
senior_df = titanic_df.loc[titanic_df["age_bin"] == "Senior"]

x = child_df["Fare"]
plt.subplot(2, 2, 1)
plt.ylabel("Number of people")
plt.title("Child Fare Price")
plt.hist(x)
x = teen_df["Fare"]
plt.subplot(2, 2, 2)
plt.ylabel("Number of people")
plt.title("Teen Fare Price")
plt.hist(x)
x = adult_df["Fare"]
plt.subplot(2, 2, 3)
plt.ylabel("Number of people")
plt.title("Adult Fare Price")
plt.hist(x)
x = senior_df["Fare"]
plt.subplot(2, 2, 4)
plt.ylabel("Number of people")
plt.title("Senior Fare Price")
plt.hist(x)
```

The teen and child fare prices generally look more spread out compared to the adults and seniors. There are a lot of children whose fare prices were much cheaper than other age groups. There were a lot of adults whose fare prices were the cheapest. It looks harder to determine if there were a lot adults who had expensive tickets because there were many adults that had cheap tickets.
