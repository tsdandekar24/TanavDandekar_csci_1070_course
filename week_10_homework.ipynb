# #1
Inductive reasoning is using specific observations to create a conclusion. An example could be Qdoba has been busy on Monday and Tuesday, therfore Qdoba will be busy on Wednesday. Deductive reasoning is when using general statements to create a specific theory. An example of deductive reasoning is all of the people in a club are at least 21 years old. A friend of yours is in the club. Therefore, your friend is at least 21 years old.

```python
combined_df = pd.concat([cleveland_df, hungarian_df, switzerland_df, va_df], ignore_index = True)
combined_df
```

```python
combined_df["Rest_bp"] = combined_df["Rest_bp"].replace("?", np.nan)
combined_df["Exercise_induced_angina"] = combined_df["Exercise_induced_angina"].replace("?", np.nan)
combined_df["ST_depression"] = combined_df["ST_depression"].replace("?", np.nan)
combined_df["Fasting_blood_sugar"] = combined_df["Fasting_blood_sugar"].replace("?", np.nan)
combined_df["Slope"] = combined_df["Slope"].replace("?", np.nan)
combined_df["Major_vessels"] = combined_df["Major_vessels"].replace("?", np.nan)
combined_df["Thal"] = combined_df["Thal"].replace("?", np.nan)
combined_df["Max_heart_rate"] = combined_df["Max_heart_rate"].replace("?", np.nan)
combined_df["Rest_ecg"] = combined_df["Rest_ecg"].replace("?", np.nan)
combined_df["Cholesterol"] = combined_df["Cholesterol"].replace("?", np.nan)
combined_df
```

```python
# Changing all of the columns from strings to floats because of the ?'s
combined_df["Rest_bp"] = combined_df["Rest_bp"].astype(float)
rest_bp_median = combined_df[~combined_df["Rest_bp"].isna()]["Rest_bp"].median()
combined_df["Rest_bp"] = combined_df["Rest_bp"].fillna(rest_bp_median)

combined_df["Exercise_induced_angina"] = combined_df["Exercise_induced_angina"].astype(float)
eang_median = combined_df[~combined_df["Exercise_induced_angina"].isna()]["Exercise_induced_angina"].median()
combined_df["Exercise_induced_angina"] = combined_df["Exercise_induced_angina"].fillna(eang_median)

combined_df["ST_depression"] = combined_df["ST_depression"].astype(float)
st_median = combined_df[~combined_df["ST_depression"].isna()]["ST_depression"].median()
combined_df["ST_depression"] = combined_df["ST_depression"].fillna(st_median)

combined_df["Fasting_blood_sugar"] = combined_df["Fasting_blood_sugar"].astype(float)
fbs_median = combined_df[~combined_df["Fasting_blood_sugar"].isna()]["Fasting_blood_sugar"].median()
combined_df["Fasting_blood_sugar"] = combined_df["Fasting_blood_sugar"].fillna(fbs_median)

combined_df["Slope"] = combined_df["Slope"].astype(float)
slope_median = combined_df[~combined_df["Slope"].isna()]["Slope"].median()
combined_df["Slope"] = combined_df["Slope"].fillna(slope_median)

combined_df["Major_vessels"] = combined_df["Major_vessels"].astype(float)
vessels_median = combined_df[~combined_df["Major_vessels"].isna()]["Major_vessels"].median()
combined_df["Major_vessels"] = combined_df["Major_vessels"].fillna(vessels_median)

combined_df["Thal"] = combined_df["Thal"].astype(float)
thal_median = combined_df[~combined_df["Thal"].isna()]["Thal"].median()
combined_df["Thal"] = combined_df["Thal"].fillna(thal_median)

combined_df["Max_heart_rate"] = combined_df["Max_heart_rate"].astype(float)
max_rate_median = combined_df[~combined_df["Max_heart_rate"].isna()]["Thal"].median()
combined_df["Max_heart_rate"] = combined_df["Max_heart_rate"].fillna(max_rate_median)

combined_df["Diagnosis"] = combined_df["Diagnosis"].astype(float)
diagnosis_median = combined_df[~combined_df["Diagnosis"].isna()]["Diagnosis"].median()
combined_df["Diagnosis"] = combined_df["Diagnosis"].fillna(diagnosis_median)

combined_df["Chest_pain"] = combined_df["Chest_pain"].astype(float)
chest_median = combined_df[~combined_df["Chest_pain"].isna()]["Chest_pain"].median()
combined_df["Chest_pain"] = combined_df["Chest_pain"].fillna(chest_median)

combined_df["Rest_ecg"] = combined_df["Rest_ecg"].astype(float)
ecg_median = combined_df[~combined_df["Rest_ecg"].isna()]["Rest_ecg"].median()
combined_df["Rest_ecg"] = combined_df["Rest_ecg"].fillna(ecg_median)

combined_df["Cholesterol"] = combined_df["Cholesterol"].astype(float)
cholesterol_median = combined_df[~combined_df["Cholesterol"].isna()]["Cholesterol"].median()
combined_df["Cholesterol"] = combined_df["Cholesterol"].fillna(cholesterol_median)

# Making the diagnosis score binary
combined_df["Diagnosis"] = np.where(combined_df["Diagnosis"] > 1, 1, combined_df["Diagnosis"])

combined_df.head(10)
```

# #3

```python
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import train_test_split

X = combined_df.drop("Diagnosis", axis = 1)
y = combined_df["Diagnosis"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
def change_model(depth, state, weight):
    model = tree.DecisionTreeClassifier(max_depth = depth, random_state = state, class_weight = weight)
change_model(8, 42, "balanced")
```

```python
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(model, feature_names = list(X.columns), filled = True, fontsize = 6, rounded = True)
plt.show()
plt.savefig("tree.svg")
```

I messed around with the max_depth to see what would bring recall and precision up. Max_depth being 8 ended up being ideal. I tried using max_features, but I found it didn't really help. It would increase my precision, but my recall ended up getting worse.

# #4

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 73, random_state = 42)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Method to adjust the rf_score
def rf_score(estimators, state):
    rf = RandomForestClassifier(n_estimators = estimators, random_state = state)
    rf = rf.fit(X_train, y_train)
    return rf.score(X_test, y_test)
rf_score(73, 42)
```

```python
ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
```

```python
rf_tree = rf.estimators_[4]
tree.plot_tree(rf_tree)
plt.savefig("rf_tree.svg")
```

I tuned the data model by changing the n_estimators. 72 and 73 ended up producing the same rf.score. I tinkered with the criterion = "log_loss" and criterion = "entropy", but it didn't changed the rf.score. I found max_features also lessened my rf.score.

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 3, n_estimators = 55, booster = "gbtree")
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Method to modify the precision and recall
def xgb_report(depth, estimators):
    xgb = XGBClassifier(max_depth = depth, n_estimators = estimators)
    xgb.fit(X_train, y_train)
    y_pred = xgb.predict(X_test)
    print(classification_report(y_test, y_pred))
xgb_report(3, 55)
```

```python
ConfusionMatrixDisplay.from_estimator(xgb, X_test, y_test)
```

I tuned the model by changing my max_depth. I found that it increased my precision by 2, but I had to sacrifice my recall by 2. I modified my n_estimators to increase my precision and recall slightly. I used different things for booster like gbtree, dart, and gblinear. I found gbtree and dart didn't really effect my precision, but gb linear reduced my precision and recall.
